```python
from pathlib import Path
import pandas as pd

mask_csv_path = "masks_subset.csv"
mask_dir = Path("/Users/mbouchou/Downloads/airbus_new_subset/images")

# Load CSV
df_masks = pd.read_csv(mask_csv_path)

# Adjust column name if needed
csv_names = set(
    df_masks["ImageId"]      # or "mask_name"
    .astype(str)
    .str.strip()
    .str.lower()
)

# Read filenames from mask folder
folder_names = {
    p.name.lower()
    for p in mask_dir.iterdir()
    if p.is_file()
}

missing_masks = csv_names - folder_names     # in CSV, not on disk
extra_masks = folder_names - csv_names       # on disk, not in CSV

print("CSV rows:", len(df_masks))
print("Unique mask names in CSV:", len(csv_names))
print("Masks in folder:", len(folder_names))
print("❌ Missing mask files:", len(missing_masks))
print("⚠️ Extra masks in folder:", len(extra_masks))
```
